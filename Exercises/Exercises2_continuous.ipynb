```python
#Task 1

#b)

def normalGenerator(nrSamples,mu,sigma2):
    #First we do Box Mueller
    U2 = np.random.uniform(0,2*np.pi,nrSamples)
    U1 = np.random.uniform(0,1,nrSamples)
    [Z1,Z2] = np.sqrt(-2*np.log(U1))*np.array([np.cos(U2),np.sin(U2)])

    N = mu + sigma2*Z1

    return N

mu = 5
sigma2 = 50

plt.hist(normalGenerator(1000000,mu,sigma2),bins=500)
```

```python
#Task 3)

"""

We use the law of large numbers to say that, the mean and the expectation both follow a normal distribution for large n

"""

mu_cis = []
sigma2_cis = []

for _ in range(100):
    N_samples = normalGenerator(10,mu,sigma2)
    mu_estimate = np.mean(N_samples)
    sigma2_estimate = np.var(N_samples)
    T = sstats.t(9)
    CI_constant = T.ppf(0.025)*np.sqrt(sigma2_estimate/10)
    mu_cis += [(mu_estimate - CI_constant,mu_estimate +CI_constant)]

```

```python
import matplotlib.pyplot as plt

# True value
true_value = mu

# Plotting the confidence intervals
fig, ax = plt.subplots()

# Plot each confidence interval
for i, (upper, lower) in enumerate(mu_cis):
    ax.plot([i, i], [lower, upper], color='blue', marker='o')

# Plot the true value as a constant line
ax.axhline(true_value, color='red', linestyle='--')

# Labeling
ax.set_xlabel('Index')
ax.set_ylabel('Value')
ax.set_title('Confidence Intervals with True Value')
plt.show()
```
